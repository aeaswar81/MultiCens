```python
import json
import pickle
```

```python
f=open("../HGv1_Gene.json")
```

```python
data=json.load(f)
```

```python
target_genes=[]
```

```python
data
```

```python
for i in data:
    target_genes.extend(data[i]['target'])
```

```python
target_genes
```

```python
len(target_genes)
```

```python
with open('target_genes.pkl', 'wb') as f:
    pickle.dump(target_genes, f)
```
